```python
from bs4 import BeautifulSoup
import requests
import urllib
from urllib.request import urlopen
import pandas as pd
```

```python
#Task 2

USER_AGENT = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:65.0) Gecko/20100101 Firefox/65.0"
headers = {"user-agent": USER_AGENT}

r = requests.get("https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off", headers=headers)
print("Status code: ", r.status_code)

soup = BeautifulSoup(r.content, "html.parser")

# finds each product from the store page
containers = soup.find("div", {"class": "_1HmYoV _35HD7C"})

# print(containers)
# container = soup.find_all("div")
container = soup.find_all("div", {"class": "bhgxx2 col-12-12"})

# print(container)

# name the output file to write to local disk

out_filename = "flipcart.csv"

# header of csv file to be written
headers = "Name,Price \n"

# opens file, and writes headers
f = open(out_filename, "w+")
f.write(headers)

# loops over each product and grabs attributes about

# each product

for con in container:
    anchors = con.find_all('div', {"class": "_1-2Iqu row"})
    if anchors:
        price = con.find("div", {"class": "_1vC4OE _2rQ-NK"})
        title = con.find('div', {"class": "_3wU53n"})
        if price is None:
            c = title.text
            b = None
        else:
            b = price.text
            c = title.text
        # prints the dataset to console
        print(c)
        print(b)
        print(c.replace(",", "") + ", " + b.replace(",", ""), file=open(out_filename, "a", encoding='utf-8'))
f.close() 
# Close the file
```
